# Import Thư viện

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve
```

# Import Dataset

```python
# Đọc file CSV không dùng header mặc định
raw_df = pd.read_csv("../Data/winequality.csv", header=None)
```

```python
raw_df.head(3)
```

Dữ liệu hiện tại đang bị lỗi định dạng
- Tất cả các cột đã bị gộp vào thành một cột duy nhất vì dấu phân cách là `;`.
- Các tên cột được nằm tong cặp `" "`

```python
# Tách tên cột từ dòng đầu tiên
column_names = raw_df.iloc[0, 0].split(";")

# Lấy dữ liệu từ dòng thứ 2 và tách theo dấu ";"
data = raw_df.iloc[1:, 0].str.split(";", expand=True)
data.columns = column_names

# Tiến hành loại bỏ dấu ngoặc kép trong tên cột
data.columns = data.columns.str.replace('"', '')

# Chuyển tất cả cột sang kiểu số
df = data.apply(pd.to_numeric)
```

# Tiền xử lý dữ liệu

### Xử lý biến mục tiêu

Do sự mất cân bằng lớn trong cột `quality` và để đơn giản hóa bài toán, chuyển đổi nó thành một bài toán **phân loại nhị phân**:
*   Tạo một cột mới `quality_binary`.
*   Gán giá trị `1` cho các loại rượu có chất lượng cao (`quality` > 6).
*   Gán giá trị `0` cho các loại rượu còn lại (`quality` <= 6).

```python
df_processed = df.copy()
quality_threshold = 6 
df_processed['quality_binary'] = df_processed['quality'].apply(lambda x: 1 if x > quality_threshold else 0)
```

```python
# Đếm số lượng và tỷ lệ phần trăm của mỗi lớp
value_counts = df_processed['quality_binary'].value_counts()
value_percent = df_processed['quality_binary'].value_counts(normalize=True) * 100

# Gộp lại thành bảng
distribution = pd.DataFrame({
    'Số lượng': value_counts,
    'Tỷ lệ (%)': value_percent.map("{:.2f}%".format)
})

print(f"\n Phân phối lớp nhị phân (0: <= {quality_threshold}, 1: > {quality_threshold}):\n")
print(distribution)
```

```python
# Loại bỏ cột 'quality' gốc
df_processed = df_processed.drop('quality', axis=1)
```

### Phân tích tương quan (Correlation Analysis)
#### Heatmap hiển thị hệ số tương quan giữa các biến

```python
correlation = df_processed.corr()
```

#### Heatmap tương quan tổng thể

```python
plt.figure(figsize=(12,12))
sns.heatmap(correlation, cbar=True, square=True, fmt = '.2f', annot = True, annot_kws={'size':12}, cmap = 'Blues')
```

#### Tương quan của các thuộc tính với `quality_binary`

```python
# Hiển thị tương quan với cột target
print("\n Mối tương quan của các thuộc tính với 'quality_binary' \n")
quality_binary_corr = correlation['quality_binary'].drop('quality_binary').sort_values(ascending=False)
print(quality_binary_corr)
```

```python
# Biểu đồ cột cho tương quan với target
plt.figure(figsize=(12, 6))
quality_binary_corr.plot(kind='bar', color=sns.color_palette("coolwarm", len(quality_binary_corr)))
plt.title('Tương quan của các thuộc tính với Quality Binary (0: Bad/Medium, 1: Good)')
plt.ylabel('Hệ số tương quan Pearson')
plt.xticks(rotation=45, ha='right') # Xoay nhãn trục x cho dễ đọc
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

- **Tương quan dương (> 0)**: Thuộc tính tăng → khả năng rượu được đánh giá chất lượng **tốt** cũng tăng.
- **Tương quan âm (< 0)**: Thuộc tính tăng → khả năng rượu thuộc loại **kém/chưa tốt** tăng lên.
- **Tương quan gần 0**: Không có mối liên hệ tuyến tính rõ ràng.

### Chia tập Huấn luyện và Kiểm tra (Train/Test Split)

```python
X = df_processed.drop('quality_binary', axis=1)
y = df_processed['quality_binary']
print("\nĐã tách thành công Features (X) và Target (y - binary).")
print("Kích thước X:", X.shape)
print("Kích thước y:", y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,  # Sử dụng 25% dữ liệu cho tập test
    random_state=42, # Để đảm bảo kết quả có thể lặp lại
    stratify=y       # Giữ tỷ lệ lớp trong cả train và test
)
```

```python
print(f"Kích thước X_train: {X_train.shape}")
print(f"Kích thước X_test: {X_test.shape}")
print(f"Kích thước y_train: {y_train.shape}")
print(f"Kích thước y_test: {y_test.shape}")
```

### Chuẩn hóa dữ liệu (Feature Scaling)
- Sử dụng `StandardScaler`.

```python
scaler = StandardScaler()

# Fit và transform trên tập huấn luyện
X_train_scaled = scaler.fit_transform(X_train)

# Chỉ transform trên tập kiểm tra (dùng mean/std đã học từ train)
X_test_scaled = scaler.transform(X_test)
```

### Huấn luyện mô hình

```python
model = RandomForestClassifier(random_state=42)
# Huấn luyện mô hình trên dữ liệu train đã chuẩn hóa
model.fit(X_train_scaled, y_train)
```

### Đánh giá Mô hình (Model Evaluation)

```python
# Dự đoán nhãn lớp
y_pred = model.predict(X_test_scaled)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Độ chính xác (Accuracy): {accuracy:.4f}")
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Dự đoán Bad/Medium (0)', 'Dự đoán Good (1)'],
            yticklabels=['Thực tế Bad/Medium (0)', 'Thực tế Good (1)'])
plt.xlabel('Nhãn Dự đoán')
plt.ylabel('Nhãn Thực tế')
plt.title('Confusion matrix')
plt.show()
```

```python
print("\n Báo cáo phân loại: \n")
target_names_binary = ['Bad/Medium (0)', 'Good (1)']
print(classification_report(y_test, y_pred, target_names=target_names_binary))
print("Giải thích:")
print(" - precision: Trong số các dự đoán cho lớp này, bao nhiêu là đúng.")
print(" - recall: Trong số các mẫu thực tế của lớp này, bao nhiêu được phát hiện đúng.")
print(" - f1-score: Trung bình điều hòa của precision và recall.")
print(" - support: Số lượng mẫu thực tế của lớp này trong tập test.")
```

### Dự đoán trên Dữ liệu mới (Prediction on New Data)

```python
input_data = (7.5,0.5,0.36,6.1,0.071,17.0,102.0,0.9978,3.35,0.8,10.5)

# Chuyển input thành mảng NumPy, reshape để thành dạng (1, số đặc trưng)
input = np.asarray(input_data).reshape(1,-1)

prediction = model.predict(input)
print(prediction)

if (prediction[0]==1):
  print('Good Quality Wine')
else:
  print('Bad Quality Wine')
```
