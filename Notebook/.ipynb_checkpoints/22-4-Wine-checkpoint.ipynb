# import Thư viện

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

# Import Dataset

```python
# Đọc file CSV không dùng header mặc định
raw_df = pd.read_csv("../Data/winequality.csv", header=None)
```

```python
raw_df.head(3)
```

Dữ liệu hiện tại đang bị lỗi định dạng
- Tất cả các cột đã bị gộp vào thành một cột duy nhất vì dấu phân cách là `;` chứ không phải dấu `,`.
- Các tên cột được nằm tong cặp `" "`

```python
# Tách tên cột từ dòng đầu tiên
column_names = raw_df.iloc[0, 0].split(";")

# Lấy dữ liệu từ dòng thứ 2 và tách theo dấu ";"
data = raw_df.iloc[1:, 0].str.split(";", expand=True)
data.columns = column_names

# Tiến hành loại bỏ dấu ngoặc kép trong tên cột
data.columns = data.columns.str.replace('"', '')

# Chuyển tất cả cột sang kiểu số
df = data.apply(pd.to_numeric)
```

### Xử lý biến mục tiêu

Do sự mất cân bằng lớn trong cột `quality` và để đơn giản hóa bài toán, chúng ta sẽ chuyển đổi nó thành một bài toán **phân loại nhị phân**:
*   Tạo một cột mới `quality_binary`.
*   Gán giá trị `1` cho các loại rượu có chất lượng cao (`quality` > 6).
*   Gán giá trị `0` cho các loại rượu còn lại (`quality` <= 6).

```python
df_processed = df.copy()
quality_threshold = 6 
df_processed['quality_binary'] = df_processed['quality'].apply(lambda x: 1 if x > quality_threshold else 0)
```

```python
# Đếm số lượng và tỷ lệ phần trăm của mỗi lớp
value_counts = df_processed['quality_binary'].value_counts()
value_percent = df_processed['quality_binary'].value_counts(normalize=True) * 100

# Gộp lại thành bảng
distribution = pd.DataFrame({
    'Số lượng': value_counts,
    'Tỷ lệ (%)': value_percent.map("{:.2f}%".format)
})

print(f"\n Phân phối lớp nhị phân (0: <= {quality_threshold}, 1: > {quality_threshold}):\n")
print(distribution)
```

```python
# Loại bỏ cột 'quality' gốc
df_processed = df_processed.drop('quality', axis=1)
```
